```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay, accuracy_score
```

```python
data = pd.read_csv("Diabetes_data.csv")
```

```python
data
```

```python
data.info()
```

```python
for gender in data["gender"].value_counts().keys():
    data['gender_' + gender] = (data["gender"] == gender).astype("int8")
data.drop("gender", axis=1, inplace=True)
```

```python
data.ethnicity.value_counts()
```

```python
for ethnicity in data.ethnicity.value_counts().keys():
    data['ethnicity_' + ethnicity] = (data['ethnicity'] == ethnicity).astype('int8')
data.drop('ethnicity', axis=True, inplace=True)
```

```python
data.education_level.value_counts()
```

```python
for edu_level in data.education_level.value_counts().keys():
    data[edu_level] = (data['education_level'] == edu_level).astype("int8")
data.drop('education_level', axis=1, inplace=True)
```

```python
data.income_level.value_counts()
```

```python
for income_level in data.income_level.value_counts().keys():
    data['income_level_' + income_level] = (data['income_level'] == income_level).astype('int8')
data.drop('income_level', axis=1, inplace=True)
```

```python
data.smoking_status.value_counts()
```

```python
for smoking_status in data.smoking_status.value_counts().keys():
    data['smoking_status_' + smoking_status] = (data['smoking_status'] == smoking_status).astype('int8')
data.drop('smoking_status', axis=1, inplace=True)
```

```python
data.employment_status.value_counts()
```

```python
for emp_status in data.employment_status.value_counts().keys():
    data['emp_status_' + emp_status] = (data['employment_status'] == emp_status).astype('int8')
data.drop('employment_status', axis=1, inplace=True)
```

```python
data.info()
```

```python
y = data['diagnosed_diabetes']
x = data.drop('diagnosed_diabetes', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
scaler = MinMaxScaler()
x_train_sc = pd.DataFrame(scaler.fit_transform(x_train), columns = x_train.columns, index = x_train.index)
x_test_sc = pd.DataFrame(scaler.transform(x_test), columns = x_test.columns, index = x_test.index)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train_sc, y_train)
```

```python
lr_pred_train = lr_model.predict(x_train_sc)
lr_pred_test = lr_model.predict(x_test_sc)
```

```python
accuracy_score(y_train, lr_pred_train), accuracy_score(y_test, lr_pred_test)
```

```python
print(classification_report(y_test, lr_pred_test))
```

```python
print(classification_report(y_train, lr_pred_train))
```

```python
NB_model = GaussianNB()
NB_model.fit(x_train_sc, y_train)
```

```python
NB_pred_train = NB_model.predict(x_train_sc)
```

```python
print(classification_report(y_train, NB_pred_train))
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train_sc, y_train)
```

```python
dt_pred_test = dt_model.predict(x_test_sc)
```

```python
print(classification_report(y_test, dt_pred_test))
```

```python
rf_model = RandomForestClassifier(n_estimators=100,random_state=42,n_jobs=-1)
rf_model.fit(x_train_sc, y_train)
```

```python
rf_pred_test = rf_model.predict(x_test_sc)
print(classification_report(y_test, rf_pred_test))
```
